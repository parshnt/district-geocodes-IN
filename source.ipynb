```python
#import required libs
import pandas as pd
from geopy.geocoders import Nominatim
```

```python
#district list
url = 'https://api.covid19india.org/csv/latest/district_list.csv'

#initialize Nominatim geocoder
geolocator = Nominatim(user_agent="your_agent") #replace your_agent

#set country for accurate results
country ="India"

#load the district list into df
df = pd.read_csv(url)
df.head()
```

```python
#create a new dataframe to hold result
new_df = pd.DataFrame( columns =['District', 'State', 'Latitude', 'Longitude']) 
```

```python
for district,state in zip(df['District'],df['State']):
    
    try:
        
        #build the query
        loc = geolocator.geocode(district+', '+state+', '+ country)
        
        #print('\nLOCATION FOUND\n'+district+', '+state,'\n',loc.latitude,',',loc.longitude)
        
        #store the incoming data inside a dict.
        data = { 'District': district, 'State':state, 'Latitude': loc.latitude, 'Longitude': loc.longitude }
        
        #append the dict. to new_df
        new_df=new_df.append(data,ignore_index=True,sort=False)
                                            
        
    except:  
        
        #print when location not found
        print("\nLOCATION NOT FOUND\n"+district+', '+state)
```

```python
#save the df to csv
new_df.to_csv('coordinates.csv') 
```
